# Use variable definition to save an `nx` graph type

Similar to creating empty lists to be populated with some sort of data such as in a list comprehension, we want to store our graph as a specific, empty type - `nx.Graph()`

**This uses the same three-pice recipe for variable assignment just like before!**

```python
nx_example = nx.Graph()
print(type(nx_example))
```

# Define the edges between nodes within the graph

We can define three edges by using the `.add_edges_from` method and specifying three relationships:

```python
nx_example.add_edges_from([
    (0,1),
    (0,2),
    (0,3)
])
```

```python
# View the nodes
nx_example.nodes()
```

```python
# view the edges
nx_example.edges()
```

or

```python
print(nx_example.nodes())
print(nx_example.edges())
```
